```python
import re
if re.search('.+@.+', 'malan@harvard.edu'):
    print('1Valid')
else:
    print("1Invalid")

if re.search('.+@.+', 'malan@harvard'):
    print('2Valid')
else:
    print("2Invalid")

if re.search('.+@.+.edu', 'malan@harvard'):
    print('3Valid')
else:
    print("3Invalid")


if re.search('.+@.+.edu', 'malan@.hhhedu'):
    print('4Valid')
else:
    print("4Invalid")

# raw string; backslash
if re.search(r'.+@.+\.edu', 'malan@2.edu333'):
    print('5Valid')
else:
    print("5Invalid")

if re.search(r'^.+@.+\.edu$', 'malan@@@malan.edu'):
    print('6Valid')
else:
    print("6Invalid")

if re.search(r'^[^@]+@[^@]+\.edu$', 'malan@@@malan.edu'):
    print('7Valid')
else:
    print("7Invalid")


```

```python
# raw string; backslash
if re.search(r'.+@.+\.edu', 'malan@2.edu333'):
    print('5Valid')
else:
    print("5Invalid")

if re.search(r'.+@.+\.edu', 'malan@2.EDU', re.IGNORECASE):
    print('5.2 Valid')
else:
    print("5.2 Invalid")
```

```python
# word character
if re.search(r'^[a-zA-Z0-9_]+@[a-zA-Z0-9_]+\.edu$', 'malan@@@malan.edu'):
    print('8.1Valid')
else:
    print("8.1Invalid")

# charactor classes  alpha number underscore
# \w any word character  \W not word 
# \s space \S not space
# \d decimal \D not decimal 
if re.search(r'^\w+@\w+\.(edu|com|org)$', 'malan@@@malan.edu'):
    print('8.2Valid')
else:
    print("8.2Invalid")

if re.search(r'^(\w|\s)', 'hi friday'):
    print('8.3 word or space Valid')
else:
    print("8.3Invalid")

if re.search(r'^\w+@\w+\.(edu|com|org)$', 'malan@cs50.harvard.edu'):
    print('8.4 Valid')
elif re.search(r'^\w+@(\w|.)+\.(edu|com|org)$', 'malan@cs50.harvard.edu'):
    print("8.4 group match")
else:
    pass

# ? 0 or 1  repetition
if re.search(r'^\w+@(\.?\w)+', 'malan@cs50.harvard.edu'):
    print('8.5 Valid')
```

```python
# format capturing groups
import re
name1 = 'Malan,          David'
# :=  walrus operator
if matches := re.search(r"^(.+), *(.+)", name1):
# if matches:
    last, first = matches.groups()
    name1 = f"{first} {last}"
print(f"hello, {name1}")

```

```python
# re.sub(pattern, repl, strings, counts, flags)  substitution

url = 'https://twitter.com/davidjmalan'
username = re.sub(r"https://twitter.com/", "", url)
print(f"username1: {username}")

url2 = 'http://www.twitter.com/davidjmalan'
username2 = re.sub(r"^(https?://)?(www\.)?twitter\.com/", "", url2)
print(f"username2: {username2}")

url3 = '  twitter.com/davidjmalan'
username3 = re.sub(r"^(https?://)?(www\.)?twitter\.com/", "", url2)
print(f"username3: {username3}")

url4 = 'http://www.twitter.com/davidjmalan'
if matches := re.search(r"^https?://(?:www\.)?twitter\.com/(.+)$", url4, re.IGNORECASE):
    print(f"username4: ", matches.group(1))
```
